#### Define hypothesis and cost function

```python
def hypothesis(x,w,b):
    return x*w + b
```

```python
def cost_mse(X,y,w,b):
    predictions = hypothesis(X,w,b)
    cost = np.mean((predictions - y)**2)
    return cost
```

#### Train with gradient descent

```python
def gradient_descent(X,y,w,b,lr = 0.01 , epochs = 1000):
    costs = []

    for i in range(epochs):
        predictions = hypothesis(X,w,b)

        dw = np.mean(2*(predictions -y) * X)
        db = np.mean(2*(predictions -y))

        w -= dw*lr
        b -= db*lr

        if i % 100 == 0:
            cost = cost_mse(X,y,w,b)
            costs.append(cost)

    return w,b,costs
```
